```python
port_stem = PorterStemmer()
stopwords = set(nltk.corpus.stopwords.words('english'))
#prepare tools for text cleaning
```

```python
def stemming(content):
  #remove non-alphabet characters and lowercase
    stemmed_content = re.sub('[^a-zA-Z]',' ',content).lower().split()
   #remove stopwords and stem
    content = [port_stem.stem(word)for word in content if word not in stopwords]
    return ' '.join(content)
```

```python
from tqdm import tqdm #tqdm='taqaddum'(Arabic word meaning 'progress')
tqdm.pandas()
new_dataset['content'] = new_dataset['content'].progress_apply(stemming)

```
